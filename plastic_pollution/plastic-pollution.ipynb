```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-01-26/plastics.csv')
```

```python
data.loc[((data["parent_company"] == "Grand Total") & (data["country"] != "EMPTY")) , ["country", "grand_total"]].sort_values(["grand_total"], ascending=False).head(10).plot(kind='bar', x="country")
plt.legend("")
plt.title("")
plt.show()
```
